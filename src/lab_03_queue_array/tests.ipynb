-   [ ] **Setup**: Copy the `Priority_Queue_array.py` template into your `login_data_structures` project.

-   [ ] **Implement `insert` and `peek`**:
    -   [ ] `insert(value)`
        -   *Hint: The `_first` attribute tracks the index of the highest priority item. Remember to update it if a new item has a higher priority.*
    -   [ ] `peek()`
        -   *Hint: This method should be simple as it just uses the `_first` attribute.*

-   [ ] **Implement `remove`**:
    -   *Hint: This is the most complex method. You cannot use built-in sort functions. Find the highest priority item (using `_first`), remove it, and then find the next highest priority item to update `_first`.*
    -   *Hint: Use the private helper method `_set_first()` to find the index of the new highest priority item after a removal.*
